```python
import random
n = 10
doors = [[]]
del doors[0]
#list2 = [["X","0","0"], ["0","X","0"], ["0","0","X"]]
for i in range(n):
    ran = random.choice("012")
    if ran == "0":
        doors.append(["X","0","0"])
    elif ran == "1":
        doors.append( ["0","X","0"])
    else:
        doors.append(["0","0","X"])
    
print(doors)
```

```python
def chanceToWin(list1):
    sum0 = 0
    sum1 = 0
    sum2 = 0
    
    for i in range(len(list1)):
        for j in range(len(list1[i])):
#            print("i:", i, "list1[i]:", list1[i][j])
            
            if j==0 and list1[i][j]=="X":
                sum0 += 1
            elif  j==1 and list1[i][j]=="X":
                sum1 += 1
            elif j==2 and list1[i][j]=="X":    
                sum2 += 1
            
    print("Sum0: ",sum0,"chance: ", sum0/len(list1))
    print("Sum1: ",sum1,"chance: ", sum1/len(list1))
    print("Sum2: ",sum2,"chance: ", sum2/len(list1))
```

```python
print(doors)


chanceToWin(doors)
```

```python
def chanceToWinIfStay(doors):
    wins = 0
    d0 = 0
    d1 = 0
    d2 = 0
    for i in range(len(doors)):
        d = doors[i].index("X")
#        if doors[i][0] == "X":
        if d == 0:
            wins += 1
#            d = doors[i].find("X")
            print("game_id: ", i, "which_door: ", d, "Win? 1")
        elif d == 1:
            d1 += 1
            print("game_id: ", i, "which_door: ", d, "Win? 0")
        else:
            d2 += 1
            print("game_id: ", i, "which_door: ", d, "Win? 0")
    print("Total: ", wins)
    print("Chance to win if stay:", wins/len(doors))
    print("Sum1: ", d1, "chance", d1/len(doors))
    print("Sum2: ", d2, "chance", d2/len(doors))
```

```python
chanceToWinIfStay(doors)
```

```python
def chance_to_win_if_switch(doors):
    wins = 0
    d0 = 0
    d1 = 0
    d2 = 0
    for i in range(len(doors)):
        d = doors[i].index("X")
#        if doors[i][0] == "X":
        if d == 0:
            d0 += 1
#            d = doors[i].find("X")
            print("game_id: ", i, "which_door: ", d, "Win? 0")
        elif d == 1:
            d1 += 1
            wins += 1
            print("game_id: ", i, "which_door: ", d, "Win? 1")
        else:
            d2 += 1
            wins += 1
            print("game_id: ", i, "which_door: ", d, "Win? 1")
    print("Total: ", len(doors))
    print("Chance to win if switch:", wins/len(doors))
    print("Chance to win if switch:", (len(doors)-wins)/len(doors))
#    print("Sum1: ", d1, "chance", d1/len(doors))
#    print("Sum2: ", d2, "chance", d2/len(doors))
```

```python
chance_to_win_if_switch(doors)
```
