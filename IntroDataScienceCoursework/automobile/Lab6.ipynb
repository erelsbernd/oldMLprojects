```python
beer_Asia = beer[beer.continent=="Asia"].beer_servings #groupby("continent")
beer_NorthAmerica = beer[beer.continent=="North America"].beer_servings
beer_Europe = beer[beer.continent=="Europe"].beer_servings
beer_Africa = beer[beer.continent=="Africa"].beer_servings
beer_Oceania = beer[beer.continent=="Oceania"].beer_servings
beer_SouthAmerica = beer[beer.continent=="South America"].beer_servings
fig = plt.figure()
fig.suptitle('LAS350A Lab 6', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)
ax.boxplot((beer_Asia, beer_NorthAmerica, beer_Europe, beer_Africa, beer_SouthAmerica, beer_Oceania))
```
