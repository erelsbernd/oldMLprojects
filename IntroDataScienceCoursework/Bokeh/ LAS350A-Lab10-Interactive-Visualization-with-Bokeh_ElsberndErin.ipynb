```python
from bokeh.charts import Bar, output_file, show
from bokeh.charts import BoxPlot, output_file, show
from bokeh.charts import Scatter, output_file, show
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
#import warnings
import numpy as np
```

```python
auto = pd.read_csv("autor2.csv")
auto.head(5)
```

```python
auto.dtypes
```

```python
p = BoxPlot(auto, values = 'price', label = ['make','engine-size'],title="Engine Size by make", color='make', plot_width=1000, plot_height=800)
output_file("boxplot_es.html")
show(p)
```

Make a scatter plot title="LAS350A - Lab 10: Price vs City-mpg (shaded by Make, maker by Bodystyle)"
with tooltips Price,Make,Bodystyle,Citympg according to the following sample:

```python
auto['bodystyle'] = auto['body-style']
auto['citympg'] = auto['city-mpg']
auto.head()
```

```python
auto['bodystyle'] = auto['body-style']
auto['citympg'] = auto['city-mpg']
auto.head()


tooltips=[
    ('Price', '@price'),
    ('Make','@make'),
    ('Body Style', '@bodystyle'),
    ('City MPG', '@citympg')
]

p = Scatter(auto, x='city-mpg', y= 'price', color='make', marker = 'body-style', title="Price vs City-mpg", plot_width=1000, plot_height=800,\
 xlabel='city-mpg', ylabel='price', tooltips= tooltips)


output_file("scatter_pm.html")
show(p)
```
